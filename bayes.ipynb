### Variabili aleatorie discrete

Una <b>variabile aleatoria discreta</b> è una coppia $(X,Pr)$ dove:\
$X$ è un insieme finito o numerabile\
$Pr : X \rightarrow [0 , 1]\quad$ soddisfa $\quad\sum_{x \in X} Pr(x) = 1$

#### Esempio (lancio di una moneta bilanciata)
$X = {0, 1}$\
$Pr(0) = Pr(1) = 1/2$

### Evento
Data una variabile aleatoria discreta $(X,Pr)$, un <b>evento</b> è un sottoinsieme di $X : E \in X$ \
La probabilità di un evento è $Pr(E) = \sum_{x \in E} Pr(x)$

#### Esempio (lancio di 3 monete bilanciate)
$X = \{0, 1\}^3$ \
$Pr(x) = (1/2)^3 = 1/8$ 

Evento = "Sono usciti più $1$ che $0$" \
$E = \{011, 101, 110, 111\}$ \
$Pr(E) = \sum_{x \in E} Pr(x) = 4 * 1/8 = 1/2$

### Probabilità congiunta
Siano $(X,Pr)$ e $(Y,Pr)$ due variabili aleatorie discrete. 

La <b>probabilità congiunta</b> $Pr(X=x, Y=y)$ è la probabilità che $X=x$ e $Y=y$.

### Indipendenza statistica 
$X$ e $Y$ sono <b>indipendenti</b> se: $\forall x \forall y \quad Pr(X=x, Y=y) = Pr(X=x)Pr(Y=y)$\
In altre parole, quando la probabilità congiunta è uguale al prodotto delle probabilità a priori.

#### Esempio (3 lanci di una moneta bilanciata)
$Y$ = esito del primo lancio\
$Y = \{0, 1\}$

$X$ = somma dei 3 lanci\
$X = \{0, 1, 2, 3\}$

Tutti i possibili risultati: $\{000, 001, 010, 011, 100, 101, 110, 111\}$\
$Y=0 : \{000, 001, 010, 011\} \quad$ <i>il primo lancio è 0</i>\
$Y=1 : \{100, 101, 110, 111\} \quad$ <i>il primo lancio è 1</i>

$Pr(Y=0) = Pr(Y=1) = 1/2$

$Pr(X=0) = 1/8$\
$Pr(X=1) = 3/8$\
$Pr(X=2) = 3/8$\
$Pr(X=3) = 1/8$

<b>$X$ e $Y$ sono indipendenti?</b>\
$\forall x \forall y : \quad Pr(X=x, Y=y) = Pr(X=x)Pr(Y=y)$

Per dimostrare che non lo sono (il contrario): $\exists x \exists y : \quad Pr(X=x, Y=y) \neq Pr(X=x)Pr(Y=y)$

$Y = 1 \quad$ <i>il primo lancio sia 1</i>\
$X = 0 \quad$ <i>la somma di tutti i lanci sia 0</i>

$Pr(X=0, Y=1) = 0 \quad$ <i>è impossibile</i>\
$Pr(X=0)Pr(Y=1) = 1/8 * 1/2 = 1/16$

$Pr(X=0, Y=1) \neq Pr(X=0)Pr(Y=1)$\
Quindi $X$ e $Y$ <b>non</b> sono indipendenti.

### Probabilità condizionata
Siano $(X,Pr)$ e $(Y,Pr)$ due variabili aleatorie discrete.

La <b>probabilità condizionata</b> $Pr(X=x | Y=y)$ è la probabilità di avere $X=x$ sapendo che $Y=y$.

Probabilità congiunta: $Pr(y,x) = Pr(y)Pr(x|y) = Pr(x,y) = Pr(x)Pr(y|x)$

### Teorema di Bayes
$Pr(y,x) = Pr(x,y)$ \
$Pr(y)Pr(x|y) = Pr(x)Pr(y|x)$

Risolvendo per $Pr(x|y)$ si ottiene il <b>teorema di Bayes</b>:

$Pr(x|y) = \frac{Pr(x)Pr(y|x)}{Pr(y)} \quad$ con $Pr(y) \neq 0$ 

### Se X e Y sono indipendenti

$Pr(x,y) = Pr(x)Pr(y|x) \quad $ [1] <i>probabilità congiunta</i>

Se $x$ e $y$ sono <b>indipendenti</b>: $\quad Pr(x,y) = Pr(x)Pr(y) \quad$ [2]

Eguagliando la formula [1] e la formula [2], otteniamo:

$Pr(x)Pr(y) = Pr(x)Pr(y|x)$\
$Pr(y) = Pr(y|x)$

Rifacendo i conti con $Pr(y,x)$ otteniamo:\
$Pr(x) = Pr(x|y)$ che ricorda la definizione di <b>cifrario perfetto</b>


### Esempio 1: Questo cifrario è perfetto? 

$P = \{a, b\}$\
$C = \{1, 2, 3\}$\
$K = {k_1, k_2}$

$P(a) = 3/4$ \
$P(b) = 1/4$ 

$P(k_1) = P(k_2) = 1/2 \quad$ <i>chiavi equiprobabili</i>

$E_{k_1}(a) = 1$\
$E_{k_1}(b) = 2$\
$E_{k_2}(a) = 2$\
$E_{k_2}(b) = 3$

#### Vogliamo dimostrare che non è perfetto:
$\exists x \in P \quad \exists y \in C :\quad Pr(x|y) \neq Pr(x)$ 

<i> Osserviamo che con queste encryption, il carattere del plaintext <b>'a'</b> non verrà mai cifrato in <b>'3'</b></i>

$x=a$ \
$y=3$

$Pr(a|3) = 0$ \
$Pr(a) = 3/4$

$Pr(a|3) \neq Pr(a) \rightarrow$ il cifrario <b>non</b> è perfetto.


### Esempio 2: Questo cifrario è perfetto? 

$P = \{a, b\}$\
$C = \{1, 2\}$\
$K = \{k_1, k_2\}$

$P(k_1) = P(k_2) = 1/2 \quad$ <i>chiavi equiprobabili</i>

$P(a) = 3/4$ \
$P(b) = 1/4$

$E_{k_1}(a) = 1$\
$E_{k_1}(b) = 2$\
$E_{k_2}(a) = 2$\
$E_{k_2}(b) = 1$\
<i>ricorda lo <b>XOR</b> dell' OTP che è perfetto</i>


#### Vogliamo dimostrare che è perfetto:

Dobbiamo dimostrare che $\forall x \in P, \quad \forall y \in C : \quad Pr(x|y) = Pr(x)$

$Pr(a|1) = Pr(a) = Pr(a|2)$\
$Pr(b|1) = Pr(b) = Pr(b|2)$

<i>Calcoliamo le probabilità condizionate:</i>

- $Pr(a|1)$

$Pr(a|1) = \frac{Pr(a)Pr(1|a)}{Pr(1)}$

$Pr(1|a) = \sum_{E_k(a)=1} Pr(k) = Pr(k_1) = 1/2 \quad$ <i>la probabilità di avere y=1 dato x=a è uguale alla somma delle probabilità delle chiavi quando $E_k(a)=1$</i>

$Pr(1) = \sum_{E_k(x)=1} Pr(x,k) \quad$<i> la probabilità di avere y=1 è uguale alla somma delle probabilità congiunte della chiave e del plaintext quando $E_k(x)=1$</i>

<i>Dal momento che $x$ e $k$ sono indipendenti:</i> $\quad Pr(x,k) = Pr(x)Pr(k)$

$Pr(1) = \sum_{E_k(x)=1} Pr(x)Pr(k)$

$Pr(1) = Pr(a)Pr(k_1) + Pr(b)Pr(k_2) = 3/4*1/2 + 1/4*1/2 = 1/2$

Sostituendo in $Pr(a|1)$ otteniamo:

$Pr(a|1) = \frac{1/2 Pr(a)}{1/2} = Pr(a)$


- $Pr(a|2)$ 

$Pr(a|2) = \frac{Pr(a)Pr(2|a)}{Pr(2)}$

$Pr(2|a) = Pr(k_2)$

$Pr(2) = Pr(a)Pr(k_2) + Pr(b)Pr(k_1)$

$Pr(a|2) = \frac{Pr(k_2)Pr(a)}{Pr(a)Pr(k_2)+Pr(b)Pr(k_1)} = \frac{1/2Pr(a)}{1/2Pr(a)+Pr(b)} = Pr(a)$


- $Pr(b|1)$ 

$Pr(b|1) = \frac{Pr(1|b)Pr(b)}{Pr(1)} = \frac{Pr(k_2)Pr(b)}{1/2} = Pr(b)$


- $Pr(b|2)$ 

$Pr(b|1) = \frac{Pr(2|b)Pr(b)}{Pr(2)} = \frac{Pr(k_1)Pr(b)}{1/2} = Pr(b)$

\
\
Il cifrario è <b>perfetto</b> per ogni scelta del plaintext a prescindere dalla loro probabilità. 


### Esempio 3: Questo cifrario è perfetto? 

$P = \{a, b\}$\
$C = \{1, 2\}$\
$K = \{k_1, k_2\}$

<i>adesso le chiavi <b>non</b> sono più equiprobabili:</i>\
$P(k_1) = 1/3$\
$P(k_2) = 2/3$ 

$P(a) = P(b) = 1/2 \quad$ <i>i plaintext sono equiprobabili</i>

$E_{k_1}(a) = 1$\
$E_{k_1}(b) = 2$\
$E_{k_2}(a) = 2$\
$E_{k_2}(b) = 1$

#### Vogliamo dimostrare che NON è perfetto:
$\exists x \in P, \quad \exists y \in C :  \quad Pr(x|y) \neq Pr(x)$

$Pr(a|1) = \frac{Pr(1|a)Pr(a)}{Pr(1)}$

$Pr(1) = Pr(a)Pr(k_1) + Pr(b)Pr(k_2) = 1/2(Pr(k_1)+Pr(k_2)) = 1/2$

$Pr(1|a) = Pr(k_1) = 1/3$

$Pr(a|1) = \frac{1/3Pr(a)}{1/2} = \frac{2}{3}Pr(a)$

$Pr(a|1) \neq Pr(a) \rightarrow$ il cifrario <b>non</b> è perfetto perchè le chiavi <b>non sono uniformemente distribuite</b>

### Segretezza dello shift sipher su singolo carattere

$P = C = K = Z_{26}$

$E_k(x) = x + k$\
$D_k(y) = y - k$

$Gen() \{k \leftarrow Z_{26}\} \quad$<i> le chiavi sono equiprobabili</i>

#### Dimostriamo che è perfetto:

$\forall x \in P, \quad \forall y \in C : \quad Pr(x|y) = Pr(x)$

$Pr(x|y) = \frac{Pr(x)Pr(y|x)}{Pr(y)} \quad$ <i>(per il teorema di Bayes)</i>

$Pr(y|x) = \sum_{E_k(x)=y} Pr(k) = \frac{1}{26} \quad$ <i>soltanto una chiave mi da $y$ come ciphertext con $x$ in input</i>

$Pr(y) = \sum_{E_k(x)=y} Pr(x,k) = \sum_{E_k(x)=y} Pr(x)Pr(k)\quad \quad$ <i>$x$ e $k$ sono <b>indipendenti</b></i>

$\sum_{E_k(x)=y} Pr(x)Pr(k) = \frac{1}{26} \sum_{E_k(x)=y} Pr(x) = \frac{1}{26} \sum_{x \in Z26} Pr(x) = \frac{1}{26} * 1 = \frac{1}{26} \quad$ <i>somma di 1/26 per 26 volte = 1</i>

$Pr(x|y) = \frac{1/26 Pr(x)}{1/26} = Pr(x)$

Quindi lo shift cipher su $Z_{26}$ <i>(su singolo carattere)</i> con chiavi uniformemente random è perfetto. 


### Shift cipher con chiavi non uniformi

$Pr(k) = \{1/2 \quad$ se $k = 25$ \
$\quad \quad \quad \quad 1/50 \quad$ se $k \neq 25\}$

#### Dimostriamo che non è perfetto:

$\exists x \in P, \quad \exists y \in C : \quad Pr(x|y) \neq Pr(x)$

<i>Proviamo con $x=0 \quad y=25$</i>

$Pr(x=0 | y=25) = \frac{Pr(y=25 | x=0) Pr(x=0)}{Pr(y=25)} \quad$ <i>teorema di Bayes</i>

$Pr(y=25 | x=0) = \sum_{E_k(0)=25} Pr(k) \quad = \quad Pr(k=25) = 1/2 \quad$ <i>la probabilità di avere ciphertext = 25 quando il plaintext è 0 è uguale alla somma delle probabilità delle chiavi quando l'encryption di $x=0$ con quelle chiavi è uguale a 25</i>

$Pr(y=25) = \sum_{E_k(x)=25}{Pr(x)Pr(k)} \quad$ <i>la probabilità di avere $y=25$ è uguale alla somma delle probabilità congiunte della chiave e del plaintext (che sono <b>indipendenti</b>) quando $E_k(x)=25$</i>

$\sum_{E_k(x)=25}{Pr(x)Pr(k)} \quad = Pr(x=0)Pr(k=25) + \sum_{Ek(x)=25 \wedge x \neq 0}{Pr(x)Pr(k)} \quad = \quad 1/2 Pr(x=0) + 1/50 \sum_{Ek(x)=25 \wedge x \neq 0}{Pr(x)} \quad = \quad 1/2 Pr(x=0) + 1/50 (1 - Pr(x=0))$

$1/2 Pr(x=0) + 1/50 (1 - Pr(x=0)) < 1/2 Pr(x=0) + 1/2 (1 - Pr(x=0)) \quad$ <i> sicuramente il primo membro è minore del secondo perchè $1/50$ è minore di $1/2$ </i>

$= 1/2 Pr(x = 0) + 1/2 - 1/2 Pr(x=0) = 1/2  \quad$ <i>sviluppando il prodotto precedente</i>

Quindi, lo shift cipher per un singolo bit con chiavi non uniformi <b>non è perfetto</b>.
