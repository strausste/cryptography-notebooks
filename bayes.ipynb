### Variabili aleatorie discrete

Una <b>variabile aleatoria discreta</b> è una coppia (X,Pr) dove:\
X è un insieme finito o numerabile\
Pr : X $\rightarrow$ \[0 , 1] soddisfa $\sum_{x \in X} Pr(x) = 1$

#### Esempio (lancio di una moneta bilanciata)
X = {0, 1}\
Pr(0) = Pr(1) = 1/2

### Evento
Data una variabile aleatoria discreta (X,Pr), un <b>evento</b> è un sottoinsieme di X : E $\in$ X \
La probabilità di un evento è Pr(E) = $\sum_{x \in E} Pr(x)$

#### Esempio (lancio di 3 monete bilanciate)
X = {0, 1}$^3$ \
Pr(x) = 1/2$^3$ = 1/8 

Evento = "Sono usciti più 1 che 0" \
E = {011, 101, 110, 111} \
Pr(E) = $\sum_{x \in E} Pr(x)$ = 4 * 1/8 = 1/2

### Probabilità congiunta
Siano (X,Pr) e (Y,Pr) due variabili aleatorie discrete. \
La <b>probabilità congiunta</b> Pr(X=x, Y=y) è la probabilità che X=x e Y=y.

### Indipendenza statistica 
X e Y sono <b>indipendenti</b> se: $\forall$x $\forall$y Pr(X=x, Y=y) = Pr(X=x)Pr(Y=y) \
In altre parole, quando la probabilità congiunta è uguale al prodotto delle probabilità a priori.

#### Esempio (3 lanci di una moneta bilanciata)
Y = esito del primo lancio\
Y = {0, 1}

X = somma dei 3 lanci\
X = {0, 1, 2, 3}

Tutti i possibili risultati: {000, 001, 010, 011, 100, 101, 110, 111}\
Y=0 : {000, 001, 010, 011} <i>il primo lancio è 0</i>\
Y=1 : {100, 101, 110, 111} <i>il primo lancio è 1</i>

Pr(Y=0) = Pr(Y=1) = 1/2

Pr(X=0) = 1/8\
Pr(X=1) = 3/8\
Pr(X=2) = 3/8\
Pr(X=3) = 1/8

<b>X e Y sono indipendenti?</b>\
$\forall$x $\forall$y : Pr(X=x, Y=y) = Pr(X=x)Pr(Y=y)\
Per dimostrare che non lo sono (il contrario): $\exists$x $\exists$y : Pr(X=x, Y=y) = Pr(X=x)Pr(Y=y)

Y = 1 <i>il primo lancio sia 1</i>\
X = 0 <i>la somma di tutti i lanci sia 0</i>

Pr(X=0, Y=1) = 0 <i>è impossibile</i>\
Pr(X=0)Pr(Y=1) = 1/8 * 1/2 = 1/16

Pr(X=0, Y=1) $\neq$ Pr(X=0)Pr(Y=1)\
Quindi X e Y <b>non</b> sono indipendenti.

### Probabilità condizionata
Siano (X,Pr) e (Y,Pr) due variabili aleatorie discrete.\
La <b>probabilità condizionata</b> Pr(X=x | Y=y) è la probabilità di avere X=x sapendo che Y=y.

Probabilità congiunta: Pr(y,x) = Pr(y)Pr(x|y) = Pr(x,y) = Pr(x)Pr(y|x)

### Teorema di Bayes
Pr(y,x) = Pr(x,y) \
Pr(y)Pr(x|y) = Pr(x)Pr(y|x)

Risolvendo per Pr(x|y) si ottiene il <b>teorema di Bayes</b>:

Pr(x|y) = $\frac{Pr(x)Pr(y|x)}{Pr(y)}$ con Pr(y) $\neq$ 0 

### Se X e Y sono indipendenti

Pr(x,y) = Pr(x)Pr(y|x) [1] <i>probabilità congiunta</i>

Se x e y sono <b>indipendenti</b>: Pr(x,y) = Pr(x)Pr(y) [2]

Eguagliando la formula [1] e la formula [2], otteniamo:

Pr(x)Pr(y) = Pr(x)Pr(y|x)\
Pr(y) = Pr(y|x)

Rifacendo i conti con Pr(y,x) otteniamo:\
Pr(x) = Pr(x|y) che ricorda la definizione di <b>cifrario perfetto</b>


### Esempio 1: Questo cifrario è perfetto? 

P = {a, b}\
C = {1, 2, 3}\
K = {k<sub>1</sub>, k<sub>2</sub>}

P(a) = 3/4 \
P(b) = 1/4 

P(k<sub>1</sub>) = P(k<sub>2</sub>) = 1/2 <i>chiavi equiprobabili</i>

E<sub>k<sub>1</sub></sub>(a) = 1\
E<sub>k<sub>1</sub></sub>(b) = 2\
E<sub>k<sub>2</sub></sub>(a) = 2\
E<sub>k<sub>2</sub></sub>(b) = 3

#### Vogliamo dimostrare che non è perfetto:
$\exists$ x $\in$ P $\exists$ y $\in$ C : Pr(x|y) $\neq$ Pr(x) 

<i> Osserviamo che con queste encryption, il carattere del plaintext <b>'a'</b> non verrà mai cifrato in <b>'3'</b></i>

x=a \
y=3

Pr(a|3) = 0 \
Pr(a) = 3/4

Pr(a|3) $\neq$ Pr(a) $\rightarrow$ il cifrario <b>non</b> è perfetto.


### Esempio 2: Questo cifrario è perfetto? 

P = {a, b}\
C = {1, 2}\
K = {k<sub>1</sub>, k<sub>2</sub>}

P(k<sub>1</sub>) = P(k<sub>2</sub>) = 1/2 <i>chiavi equiprobabili</i>

P(a) = 3/4 \
P(b) = 1/4

E<sub>k<sub>1</sub></sub>(a) = 1\
E<sub>k<sub>1</sub></sub>(b) = 2\
E<sub>k<sub>2</sub></sub>(a) = 2\
E<sub>k<sub>2</sub></sub>(b) = 1\
<i>ricorda lo <b>XOR</b> dell' OTP che è perfetto</i>


#### Vogliamo dimostrare che è perfetto:

Dobbiamo dimostrare che $\forall$ x $\in$ P $\forall$ y $\in$ C : Pr(x|y) = Pr(x)

Pr(a|1) = Pr(a) = Pr(a|2)\
Pr(b|1) = Pr(b) = Pr(b|2)

<i>Calcoliamo le probabilità condizionate:</i>

- Pr(a|1) 

Pr(a|1) = $\frac{Pr(a)Pr(1|a)}{Pr(1)}$

Pr(1|a) = $\sum_{Ek(a)=1} Pr(k)$ = Pr(k1) = 1/2 <i>la probabilità di avere y=1 dato x=a è uguale alla somma delle probabilità delle chiavi quando E<sub>k</sub>(a)=1</i>

Pr(1) = $\sum_{Ek(x)=1} Pr(x,k)$ <i>la probabilità di avere y=1 è uguale alla somma della probabilità congiunta della chiave e del ciphertext quando E<sub>k</sub>(x)=1</i>

<i>Dal momento che x e k sono indipendenti:</i> Pr(x,k) = Pr(x)Pr(k) 

Pr(1) = $\sum_{Ek(x)=1} Pr(x)Pr(k)$

Pr(1) = Pr(a)Pr(k1) + Pr(b)Pr(k2) ) 3/4\*1/2 + 1/4\*1/2 = 1/2

Sostituendo in Pr(a|1) otteniamo:

Pr(a|1) = $\frac{1/2 Pr(a)}{1/2}$ = Pr(a) 


- Pr(a|2) 

Pr(a|2) = $\frac{Pr(a)Pr(2|a)}{Pr(2)}$

Pr(2|a) = Pr(k<sub>2</sub>)

Pr(2) = Pr(a)Pr(k<sub>2</sub>) + Pr(b)Pr(k<sub>1</sub>)

Pr(a|2) = $\frac{Pr(k2)Pr(a)}{Pr(a)Pr(k2)+Pr(b)Pr(k1)}$ = $\frac{1/2Pr(a}{1/2Pr(a)+Pr(b)}$ = Pr(a)


- Pr(b|1) 

Pr(b|1) = $\frac{Pr(1|b)Pr(b)}{Pr(1)}$ = $\frac{Pr(k2)Pr(b)}{1/2}$ = Pr(b)


- Pr(b|2) 

Pr(b|1) = $\frac{Pr(2|b)Pr(b)}{Pr(2)}$ = $\frac{Pr(k1)Pr(b)}{1/2}$ = Pr(b)

\
\
Il cifrario è <b>perfetto</b> per ogni scelta del plaintext a prescindere dalla loro probabilità.
